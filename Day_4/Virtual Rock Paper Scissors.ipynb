```python
#Rock paper scissors game
import random
seed_number = input("create a seed number")
random.seed(seed_number)

print("welcome to the virtual Rock, Paper Scissors game")
print("input 0 for rock, 1 for paper amd 2 for scissors")

computer = random.randint(0,2)

scissors =   '''_______
            ---'   ____)___
                    ______)
                __________)
                (____)
            ---.__(___) '''

paper =  '''_______
        ---'    ____)_
                ______)
                _______)
                _______)
        ---.__________)'''

rock =   '''_______
        ---'   ____)
            (_____)
            (_____)
            (____)
        ---.__(___)'''


human = int(input("what do choose??"))

if human == 0:
    print(f"You chose rock\n{rock}")
    if computer == 2:
        print(f"Computer chose scissors\n {scissors}")
        print("you win")
    elif computer == 1:
        print(f"Computer chose paper\n {paper}")
        print("you lost")
    elif computer == 0:
        print(f"Computer chose rock\n {rock}")
        print("it is a tie")
    else:
        print("WRONG INPUT!!!")
elif human == 1:
    print(f"You chose paper\n {paper}")
    if computer == 0:
        print(f"Computer chose rock\n {rock}")
        print("you win")
    elif computer == 2:
        print(f"Computer chose scissors\n {scissors}")
        print("you lost")
    elif computer == 1:
        print(f"Computer chose paper\n {paper}")
        print("it is a tie")
    else: 
        print("WRONG INPUT!!!")
elif human == 2:
    print(f"You chose scissors\n {scissors}")
    if computer == 1:
        print(f"Computer chose paper\n {paper}")
        print("you win")
    elif computer == 0:
        print(f"Computer chose rock\n {rock}")
        print("you lost")
    elif computer == 2:
        print(f"Computer chose scissors\n {scissors}")
        print("it is a tie")
    else: 
        print("WRONG INPUT!!!")
```
